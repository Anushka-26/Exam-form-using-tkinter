```python
import tkinter as tk
from tkinter import messagebox

def saved():
    g_enroll = enroll.get()
    g_user = user.get()
    g_passw = passw.get()
    g_confp = confp.get()
    g_age = age.get()

    if g_passw != g_confp :
        messagebox.showerror(title='failure',message='passwords dont match')
    elif int(g_age)<18:
        messagebox.showerror(title='failure',message='age not above 18')
    else:
        file = open('dataform.txt','a')
        file.write(g_enroll+','+g_user+','+g_passw+','+g_age+'\n')
        file.close()
        messagebox.showinfo(title='success',message='user registered')
    
    enroll.delete(0,tk.END)
    user.delete(0,tk.END)
    passw.delete(0,tk.END)
    confp.delete(0,tk.END)
    age.delete(0,tk.END)
   
    
win=tk.Tk()
win.geometry("400x400")
win.title("EXAMINATION PORTAL")
win.config(bg='green')

l1 = tk.Label(win,text='REGISTRATION FORM',bg='white',fg='black',font=('Times new roman',20))
l1.pack()

tk.Label(win,text='ENROLLMENT NUMBER',bg='white',fg='black',font=('Times new roman',8)).place(x=20,y=70)
enroll = tk.Entry(win,font=('Times new roman',11))
enroll.place(x=200,y=70)

tk.Label(win,text='USERNAME',bg='white',fg='black',font=('Times new roman',8)).place(x=20,y=100)
user = tk.Entry(win,font=('Times new roman',11))
user.place(x=200,y=100)

tk.Label(win,text='PASSWORD',bg='white',fg='black',font=('Times new roman',8)).place(x=20,y=130)
passw = tk.Entry(win,font=('Times new roman',11))
passw.place(x=200,y=130)

tk.Label(win,text='CONFIRM PASSWORD',bg='white',fg='black',font=('Times new roman',8)).place(x=20,y=160)
confp = tk.Entry(win,font=('Times new roman',11))
confp.place(x=200,y=160)

tk.Label(win,text='AGE',bg='white',fg='black',font=("Times new roman",8)).place(x=20,y=190)
age = tk.Entry(win,font=("Times new roman",11))
age.place(x=200,y=190)

tk.Button(win,text='SUBMIT',command=saved,bg='purple',fg='white',font=("Times new roman",15)).place(x=130,y=250)

win.mainloop()
```
